### Time Series Decomposition

```python
import pandas as pd
url = "https://raw.githubusercontent.com/utkarshg1/Neural-Net-Datasets/refs/heads/main/AirPassengers.csv"
df = pd.read_csv(url, index_col="Month", parse_dates=True)
df
```

#### Date format - yyyy-mm-dd

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,15))
plt.plot(df["Passengers"])
plt.xlabel("Month")
plt.ylabel("Passenger")
plt.title("AirPassengers by month")
plt.grid()
plt.show()
```

#### Additive - trend+season+Resid
#### Multiplicative - trendxSeasonxResid

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
d = seasonal_decompose(df["Passengers"], model="multiplicative")
d.plot()
```

```python
# show the plot for trend
plt.figure(figsize=(10,5))
plt.plot(df["Passengers"])
d.trend.plot()
plt.xlabel("Month")
plt.ylabel("Passenger")
plt.title("AirPassengers Trendline")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
d.seasonal.plot()

plt.xlabel("Month")
plt.ylabel("Passengers")
plt.title("Season")
plt.grid()
plt.show()
```

```python

plt.figure(figsize=(10,5))
d.seasonal.plot()

plt.xlabel("Month")
plt.ylabel("Passengers")
plt.title("Residual")
plt.grid()
plt.show()
```

#### Time Series decomposition complete
### Time Series Forecasting

#### ARIMA - Auto Regressive Integrated Moving Average

###### ARIMA(p, d, q)

```python
%pip install pmdarima
```

```python
from pmdarima.arima import auto_arima
```

```python
# consider 1 year as season 1 year = 12 months
model = auto_arima(df["Passengers"], seasonal= True, m=12)
```

```python
model.order
```

#### ARIMA(2,1,1)
#### Forecast value for next 4 years -> 12*4 = 48 months

```python
forecast, conf = model.predict(n_periods =48, return_conf_int =True)
```

```python
forecast
```

```python
conf
```

```python
df_fo = pd.DataFrame(forecast, columns=["forecast"])
df_fo.head()
```

```python
df_fo[["Low", "High"]] =conf
```

```python
df_fo = df_fo.round(0)
df_fo
```

```python
df_fo.to_csv("Forecast.csv", index_label="Month")
```

#### Visualize the Forecast

```python
plt.figure(figsize=(10, 5))
plt.plot(df["Passengers"], label="observed")
plt.fill_between(df_fo.index, df_fo["Low"], df_fo["High"], color="pink", alpha=0.9, label="confidence")
plt.plot(df_fo["forecast"], label="forecast")
d.trend.plot()
plt.xlabel("Month")
plt.ylabel("Passenger")
plt.title("AirPassengers Forecast")
plt.legend()
plt.grid()
plt.show()
```
